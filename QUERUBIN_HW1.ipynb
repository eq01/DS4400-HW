## [C] Problem 5: Average, variance, and correlation (20 points)

The dataset for this assignment is available [here](https://drive.google.com/drive/folders/1FV7qofHD_olqygrIjASLQ7k7LTyVSEGG?usp=sharing). The prediction task is to predict the price of a house (column price) given the other features. Please ignore the columns id and date, as well as the categorical column zipcode. File kc_house_data.csv includes all the records in the dataset. You should use the entire dataset for the assignment. You can also find a Word document including the feature description in the same folder. In this problem, we will perform some exploratory data analysis using the house price dataset.

1. For each feature, write code to compute the average value, the min and max values, as well as its variance.
    
    Which features have the lowest and the highest average? Include the feature name and their average values for the features with the
    lowest and highest average.
    
    Which features have the lowest and the highest variance? Include the feature name and their variance values for the features with the
    lowest and highest variance.
    
2. Compute the correlation coefficient of each feature with the response. Include a table with the correlation coefficient of each feature with the response. 
Which features are positively correlated (i.e., have positive correlation coefficient) with the response?

Which feature has the highest positive correlation with the response?

3. Were you able to find any features with a negative correlation coefficient with the response?

```python
import pandas as pd
import numpy as np

df = pd.read_csv('kc_house_data.csv')

remove = ['id', 'date', 'zipcode']
df_clean = df.drop(columns=remove)

y = df_clean["price"]
X = df_clean.drop(columns=["price"])

stats = pd.DataFrame({
    "average": X.mean(),
    "min": X.min(),
    "max": X.max(),
    "variance": X.var()
})    

# average
lowest_avg_feature = stats["average"].idxmin()
highest_avg_feature = stats["average"].idxmax()

lowest_avg_value = stats.loc[lowest_avg_feature, "average"]
highest_avg_value = stats.loc[highest_avg_feature, "average"]

print(f"\nLowest Average: {lowest_avg_feature} = {lowest_avg_value}")
print(f"Highest Average: {highest_avg_feature} = {highest_avg_value}")

# variance
lowest_var_feature = stats["variance"].idxmin()
highest_var_feature = stats["variance"].idxmax()

lowest_var_value = stats.loc[lowest_var_feature, "variance"]
highest_var_value = stats.loc[highest_var_feature, "variance"]

print(f"\nLowest Variance: {lowest_var_feature} = {lowest_var_value}")
print(f"Highest Variance: {highest_var_feature} = {highest_var_value}")
```

```python
# find correlation of x with y
correlations = X.corrwith(y)

# create corr table and sort by price
correlation_table = pd.DataFrame({
    "feature": correlations.index,
    "correlation_with_price": correlations.values
}).sort_values(by="correlation_with_price", ascending=False)

# find pos correlations
positive_correlations = correlation_table[correlation_table["correlation_with_price"] > 0]

# feature with highest pos corr
highest_positive_feature = correlation_table.iloc[0]["feature"]
highest_positive_value = correlation_table.iloc[0]["correlation_with_price"]

# fnd any negative correlations
negative_correlations = correlation_table[correlation_table["correlation_with_price"] < 0]

# print results
print("\nCorrelation with Price:")
print(correlation_table)

print("\nPositively Correlated Features:")
print(positive_correlations)

print("\nNegatively Correlated Features:")
print(negative_correlations)

print("\nHighest Positive Feature:")
print(highest_positive_feature)

print("\nHighest Positive Value:")
print(highest_positive_value)
```
